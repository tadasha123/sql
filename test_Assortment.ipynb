```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
```

```python
#1.Read in the data from the supplied csv (test_assortments.csv)
test_df=pd.read_csv('test_assortments.csv')
test_df.head()
```

```python
#2.Show null counts by column
test_df.isnull().sum()
```

```python
#3.Fill in null values in 'store_type_code' with 'RED' and in 'prediction' with 0
test_df.fillna({'store_type_code':'RED','prediction':0})
```

```python
#4. Create a new data frame from the original that contains only records(rows) with 'prediction' more than 0. 
#Use this data frame (named df) in below steps.
df=test_df[test_df['prediction']>0]
df
```

```python
#5. Add a column 'predicted_gain' with a value calculated as:  prediction * (new_order_point - old_order_point).
df['predicted_gain']=df['prediction']*df['new_order_point']-df['old_order_point']
df
```

```python
#6. Add a column 'grouped_reason' to the data frame with three values: 
    #"REMOVE" if 'reason' starts with 'REMOVE', "ADD" if it starts with 'ADD' and 'ELSE' for everything else.
df['grouped_reason']=['REMOVE'  if r.startswith('REMOVE') else 'ADD' if r.startswith('ADD') else 'ELSE' for r in df['reason']]
df
```

```python
#7. Create a dictionary (dict1) with the keys being unique values in the column 'reason' 
#and values equal to 20 (for every key(reason)). Create a copy of dict1 as dict2.

dict1 = dict((reason, 20) for reason in df['reason'].unique())
print(dict1)
dict2=dict1.copy()
```

```python
#8. Order the df by 'prediction' column (descending). 
#Iterate over the rows of the df to check if the reason is in the dictionary 
#(Note: they all should be in the dict but let's add this check).
#If the reason is in the dict2, subtract 1 from the dict2 key's value. Stop subtracting from the key's value if it reaches 0. (The iteration loop should go to the end of the df.)
df=df.sort_values('prediction',ascending='False')
for i in df['reason']:
    if i in dict2.keys() and dict2[i]>0:
        dict2[i]-=1       
dict2       
```

```python
#9. Convert the dict2 into a data frame 'df_from_dict2' so that you have two columns: 'reason' and 'remaining_count'.
df_from_dict2=pd.DataFrame(data=list(dict2.items()),columns=['reason','remaining_count'])
df_from_dict2
```

```python
#10. Select only the 'store_number', 'skunum', 'reason' from the df and join with the 'df_from_dict2' on 'reason'. 
#Show the result.
#df[['store_number','skunum','reason']]
#df['skunum']#,'reason']
pd.merge(df[['store_number','skunum','reason']],df_from_dict2,on='reason')
```
